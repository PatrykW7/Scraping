```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd


url = "https://www.carpages.ca/used-cars/search/?fueltype_id%5B0%5D=3&fueltype_id%5B1%5D=7"

page = requests.get(url)

soup = BeautifulSoup(page.text,"lxml")

df = None
df = pd.DataFrame({"Link":[],"Title":[],"Price":[]})

iterator = 0
#while iterator <= 15:

part_link = "https://www.carpages.ca/"    
boxes = soup.find_all("div",class_ = "media soft push-none rule")

while iterator < 3:
    for box in boxes:
        try:
            #link  = box.find("a",{"title":"2010 Lexus RX 450h AWD 4DR Hybrid"}).get("href")
            link = box.find_all("h4",{"class":"hN"})[0].find("a").get("href")
            full_link = part_link+link
            title = box.find("h4",{"class":"hN"}).text.strip()
            price = box.find("strong",class_ = "delta").text.strip()
            #color = box.find_all("span")[9].text.strip()
            #color = box.find_all("div",class_ = "grey l-column l-column--small-6 l-column--medium-4")[1].text.strip()
            df = df.append({"Link":full_link,"Title":title,"Price":price},ignore_index = True)
        except:

            pass
    #break
    

    next_page = soup.find("a",{"class":"nextprev"}).get("href")
    full_next_page = part_link+next_page
    full_next_page
    
    url = full_next_page
    page = requests.get(url)
    soup = BeautifulSoup(page.text,"lxml")
    iterator+=1
    
    
```
